#### Final Approach:-
##### • Scatter plot to show the relationship between GDP per capita and happiness score.
#### • Heat map to show the correlations between different variables.
##### • Bar plot to show the Happiness score for the top 10 countries.
###### • Box plot to show the distribution of each variable in the dataset.
###### • Pair Plot to show the relationship between each variable in the dataset.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('happiness_report.csv')
```

## scatter plot of GDP per capita vs. Happiness Score

```python
sns.scatterplot(x='GDP per capita', y='Score', data=df)
plt.xlabel('GDP per capita')
plt.ylabel('Happiness Score')
plt.title('GDP per capita vs. Happiness Score')
plt.show()
```

## heatmap of correlations between variables

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlations between Variables')
plt.show()
```

## bar plot of Happiness Score for top 10 countries

```python
top_10 = df.head(10)
sns.barplot(x='Country or region', y='Score', data=top_10)
plt.xlabel('Country or Region')
plt.xticks(rotation=45)
plt.ylabel('Happiness Score')
plt.title('Happiness Score for Top 10 Countries')
plt.show()
```

## box plot of variable distributions

```python
sns.boxplot(data=df)
plt.title('Distribution of Variables')
plt.xticks(rotation=45)
plt.show()
```

## pair plot of all variables

```python
sns.pairplot(data=df)
plt.suptitle('Pair Plot of All Variables',fontsize=20, y=1.05,c='red')
plt.show()
```
