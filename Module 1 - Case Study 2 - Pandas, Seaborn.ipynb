```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Checking the count of categories that have more presence on YouTube

```python
sns.countplot(x='Category', data=df)
plt.title('Count of Influencers in Each Category')
plt.xticks(rotation=90)
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```
