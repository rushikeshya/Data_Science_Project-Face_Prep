### Importing Libraries

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('website_speed.csv')
```

### Split the data into two groups based on page load time:

```python
fast = data[data['Page load time'] < data['Page load time'].median()]
slow = data[data['Page load time'] >= data['Page load time'].median()]
```
